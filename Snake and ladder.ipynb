```python
import random

def roll_die():
    return random.randint(1, 6)

def snake_ladder(position):
    snakes_and_ladders = {
        16: 6, 47: 26, 49: 11, 56: 53, 62: 19,
        64: 60, 87: 24, 93: 73, 95: 75, 98: 78
    }
    return snakes_and_ladders.get(position, position)

def print_board(player1, player2):
    board = list(range(1, 101))
    for position in [player1, player2]:
        if 1 <= position <= 100:
            board[position - 1] = f"P{board[position - 1]:02d}"

    for i in range(10):
        row = board[i * 10:(i + 1) * 10]
        print(" | ".join(map(str, row)))
        print("-" * 50)

def snake_ladder_game():
    player1, player2 = 1, 1

    while True:
        input("Player 1 - Press Enter to roll the die.")
        dice_roll = roll_die()
        print(f"Player 1 rolled a {dice_roll}.")
        player1 += dice_roll
        player1 = snake_ladder(player1)

        print_board(player1, player2)

        if player1 >= 100:
            print("Player 1 wins!")
            break

        input("Player 2 - Press Enter to roll the die.")
        dice_roll = roll_die()
        print(f"Player 2 rolled a {dice_roll}.")
        player2 += dice_roll
        player2 = snake_ladder(player2)

        print_board(player1, player2)

        if player2 >= 100:
            print("Player 2 wins!")
            break

if __name__ == "__main__":
    snake_ladder_game()
    
```
